## Modeling

### K-Means

Using K-Means to cluster

**Standardize the data**

```python
scaler = StandardScaler() 

scaler_ft = scaler.fit_transform(data_preped) 

data_scaled = pd.DataFrame(scaler_ft, columns=data_preped.columns) 
data_scaled.describe()
```

**Fiting models**

```python
ModInertia = []
ModSil=[]

for n in range(2,16):
    
    modelK = KMeans(n_clusters=n, n_init='auto', random_state=42)
    modelK.fit(data_scaled)
    
    ModInertia.append(modelK.inertia_)
    ModSil.append(silhouette_score(data_scaled, modelK.labels_, metric='euclidean', sample_size=None))
```

```python
Series_inertia = pd.Series(ModInertia, index=range(2,16))
Series_inertia.plot();
```

```python
Series_sil = pd.Series(ModSil, index=range(2,16))
Series_sil.plot();
```

- The Elbows from inertia could be 4 or 7
- the peaks of the silhouette score are 4 and 7 (4 beeing the highest)

```python
kmeans4 = KMeans(n_clusters=4, n_init='auto', random_state=42)
kmeans4.fit(data_scaled)
```

```python
kmeans7 = KMeans(n_clusters=7, n_init='auto', random_state=42)
kmeans7.fit(data_scaled)
```

```python
centers7 = kmeans7.cluster_centers_

cluster_centers7 = pd.DataFrame(centers7, columns=data_scaled.columns)

sns.heatmap(cluster_centers7, cmap='RdBu');
```

## Evaluation

Going with the 6 clusters without department data.

```python
KMeans6
```

```python
Counter(KMeans6.labels_)
```

**Cluster mapping**

```python
Clusters = pd.DataFrame(KMeans6.labels_, columns=['Cluster'])
Clusters
```

```python
Cluster_mapping = {0: 'Senior employees',
                   1: 'Juniors that live far',
                   2: 'Juniors that are satisfied',
                   3: 'Juniors that are unsatisfied',
                   4: 'Female Juniors',
                   5: 'Juniors that peform well'}
```

```python
Clusters['Cluster_Name'] = Clusters['Cluster'].map(Cluster_mapping)
Clusters
```

**Attrition rates**

```python
df_cluster_att = pd.concat([Clusters, data.Attrition], axis=1)
df_cluster_att
```

```python
df_cluster_att.groupby(['Cluster_Name'])['Attrition'].mean().sort_values(ascending=False)
```

```python
df_cluster_att.Cluster_Name.value_counts()
```

- People that live far tend to have more attrition
- Senior people tend to have less attrition

**Department analysis**

```python
df_cluster_att_depart = pd.concat([df_cluster_att, data.Department], axis=1)
df_cluster_att_depart
```

```python
df_cluster_att_depart.groupby(['Cluster_Name', 'Department']).mean().sort_values('Attrition', ascending=False)
```

- The groups most likely to leave are people that live far or in HR
- The group most likely not to leave are senior employees

```python
df_cluster_att_depart.groupby(['Cluster_Name', 'Department']).count()
```

**Recommendations**

To inclease employee retention attrition:
- For people that live far: find remote options (home office or hybrid) and create a good remote culture
- People who are unsatisfied: talk to the managers for them to understand what is happening more specifically
